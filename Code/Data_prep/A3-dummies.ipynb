```
def forzar_tipos(df, info, estandarizar=0):
  from sklearn.preprocessing import StandardScaler
  list_float = list(info.campo[info['tipo'] == 'float64'])
  list_obj = list(info.campo[info['tipo'] == 'object'])
  list_int =  list(info.campo[info['tipo'] == 'int64'])
  df1 = pd.DataFrame()
  df1[list_obj] = df[list_obj].astype(object)
  df1[list_float] = df[list_float].astype(float) 
  df1[list_int] = df[list_int].astype(int)
  if estandarizar == 1:
    df1[list_float] = StandardScaler().fit_transform(df1[list_float])
  return df1
```

##Variables dummies
Dada la presencia de variables nominales, se convierten en variables dummies. Tendrá un valor de uno cuando ocurre un evento categórico y cero cuando no ocurre.


```
A0  = pd.get_dummies(df2018['p_5'],prefix='p_5')
A1  = pd.get_dummies(df2018['p_6'],prefix='p_6')
A2  = pd.get_dummies(df2018['p_7'],prefix='p_7')
A16 = pd.get_dummies(df2018['p_10'],  prefix='p_10')
A3  = pd.get_dummies(df2018['p_146'], prefix='p_146')
A4  = pd.get_dummies(df2018['p_147'], prefix='p_147')
A5  = pd.get_dummies(df2018['p_148'], prefix='p_148')
A6  = pd.get_dummies(df2018['p_158'], prefix='p_158')
A7  = pd.get_dummies(df2018['p_162'], prefix='p_162')
A8  = pd.get_dummies(df2018['p_165'], prefix='p_165')
A9  = pd.get_dummies(df2018['p_171'], prefix='p_171')
A10 = pd.get_dummies(df2018['p_174'], prefix='p_174')
A11 = pd.get_dummies(df2018['p_178'], prefix='p_178')
A12 = pd.get_dummies(df2018['p_184'], prefix='p_184')
A13 = pd.get_dummies(df2018['p_284'], prefix='p_284')
A14 = pd.get_dummies(df2018['p_285'], prefix='p_285')
A15 = pd.get_dummies(df2018['p_286'], prefix='p_286')
```

```
D=pd.concat([A0, A1, A2, A16, A3, A4, A5, A6, A7, A8, A9, A10, A11, A12, A13, A14, A15], axis=1)
D=pd.DataFrame(D)
D.head()
```

```
D = D.astype(object)
```
